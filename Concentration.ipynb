```python
import numpy as np

# Calculate precision and recall for each class
precision_per_class = precision_score(y_test, y_pred, average=None)
recall_per_class = recall_score(y_test, y_pred, average=None)

# Calculate micro-average precision and recall
micro_precision = precision_score(y_test, y_pred, average='micro')
micro_recall = recall_score(y_test, y_pred, average='micro')

# Calculate macro-average precision and recall
macro_precision = precision_score(y_test, y_pred, average='macro')
macro_recall = recall_score(y_test, y_pred, average='macro')

# Calculate weighted-average precision and recall
weighted_precision = precision_score(y_test, y_pred, average='weighted')
weighted_recall = recall_score(y_test, y_pred, average='weighted')

# Print per-class precision and recall
num_classes = len(np.unique(y_test))
for class_label in range(num_classes):
    print(f"Class {class_label}: Precision = {precision_per_class[class_label]}, Recall = {recall_per_class[class_label]}")

print("Micro-Average Precision:", micro_precision)
print("Micro-Average Recall:", micro_recall)
print("Macro-Average Precision:", macro_precision)
print("Macro-Average Recall:", macro_recall)
print("Weighted-Average Precision:", weighted_precision)
print("Weighted-Average Recall:", weighted_recall)
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_curve, cohen_kappa_score
import numpy as np
import matplotlib.pyplot as plt

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate False Positive Rate (FPR) and False Negative Rate (FNR) for each class
num_classes = len(np.unique(y_test))
fpr_per_class = []
fnr_per_class = []
for class_label in range(num_classes):
    tn = np.sum(np.delete(cm, class_label, axis=0)[:, class_label])
    fp = np.sum(cm[:, class_label]) - cm[class_label, class_label]
    tp = cm[class_label, class_label]
    fn = np.sum(np.delete(cm, class_label, axis=1)[class_label, :])
    fpr = fp / (fp + tn)
    fnr = fn / (fn + tp)
    fpr_per_class.append(fpr)
    fnr_per_class.append(fnr)

# Calculate Cohen's Kappa
cohen_kappa = cohen_kappa_score(y_test, y_pred)

print("False Positive Rate per Class:", fpr_per_class)
print("False Negative Rate per Class:", fnr_per_class)
# print("F Beta score per Class:", f_beta_per_class)
print("Cohen's Kappa:", cohen_kappa)
```
